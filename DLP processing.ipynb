# Processing parameters

```python
import pandas as pd

df = pd.read_csv("meus_artigos.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# Defina aqui os fatores de processamento de interesse
# -------------------------------------------------------
fatores = ["slurry", "suspension", "loading", "rheology", "viscosity", "printing", "curing", "debinding", "sintering"]

# -------------------------------------------------------
# Coluna onde as keywords estão armazenadas
# -------------------------------------------------------
COLUNA_KEYWORDS = "Author Keywords"

# -------------------------------------------------------
# Para cada fator, cria uma coluna True/False no dataframe
# indicando se aquele artigo menciona o fator
# -------------------------------------------------------
for fator in fatores:
    df[fator] = df[COLUNA_KEYWORDS].str.contains(fator, case=False, na=False)

# -------------------------------------------------------
# Agrupa por ano e soma as ocorrências de cada fator
# -------------------------------------------------------
tabela = df.groupby("Year")[fatores].sum().astype(int)

# Renomeia as colunas para ficarem capitalizadas
tabela.columns = [f.capitalize() for f in fatores]

# Adiciona uma coluna com o total geral por ano
tabela["Total"] = tabela.sum(axis=1)

# Adiciona uma linha com o total geral por fator
tabela.loc["Total"] = tabela.sum()

# -------------------------------------------------------
# Exibe a tabela no terminal
# -------------------------------------------------------
print(f"📊 Ocorrências por ano em '{COLUNA_KEYWORDS}'\n")
print(tabela.to_string())

# -------------------------------------------------------
# Salva em Excel
# -------------------------------------------------------
NOME_ARQUIVO = "fatores_processamento_por_ano.xlsx"
tabela.to_excel(NOME_ARQUIVO)
print(f"\n💾 Arquivo salvo: {NOME_ARQUIVO}")
```
