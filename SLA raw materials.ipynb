# Count raw materials 

```python
import pandas as pd

df = pd.read_csv("meus_artigos_sla.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# Defina aqui as matérias primas de interesse
# -------------------------------------------------------
materiais = ["alumina", "zirconia", "silica", "hydroxyapatite", "nitride", "mullite", "silicon carbide", "silicon", "barium titanate", "phosphate", "silicate"]

# -------------------------------------------------------
# Coluna onde as keywords estão armazenadas
# -------------------------------------------------------
COLUNA_KEYWORDS = "Author Keywords"

# -------------------------------------------------------
# Para cada material, cria uma coluna True/False no dataframe
# indicando se aquele artigo menciona o material
# -------------------------------------------------------
for material in materiais:
    df[material] = df[COLUNA_KEYWORDS].str.contains(material, case=False, na=False)

# -------------------------------------------------------
# Agrupa por ano e soma as ocorrências de cada material
# O resultado é uma tabela: linhas = anos, colunas = materiais
# -------------------------------------------------------
tabela = df.groupby("Year")[materiais].sum().astype(int)

# Renomeia as colunas para ficarem capitalizadas (ex: "alumina" → "Alumina")
tabela.columns = [m.capitalize() for m in materiais]

# Adiciona uma coluna com o total geral por ano
tabela["Total"] = tabela.sum(axis=1)

# Adiciona uma linha com o total geral por material
tabela.loc["Total"] = tabela.sum()

# -------------------------------------------------------
# Exibe a tabela no terminal
# -------------------------------------------------------
print(f"📊 Ocorrências por ano em '{COLUNA_KEYWORDS}'\n")
print(tabela.to_string())

# -------------------------------------------------------
# Salva em Excel
# -------------------------------------------------------
NOME_ARQUIVO = "materiais_por_ano.xlsx"
tabela.to_excel(NOME_ARQUIVO)
print(f"\n💾 Arquivo salvo: {NOME_ARQUIVO}")
```
