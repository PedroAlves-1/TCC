# China publications

```python
import pandas as pd
from collections import Counter

df = pd.read_csv("meus_artigos.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# País de interesse e intervalo de anos
# -------------------------------------------------------
PAIS       = "China"
ANO_INICIO = 2000
ANO_FIM    = 2025

# -------------------------------------------------------
# Coluna onde as afiliações estão armazenadas
# O país aparece como último elemento de cada afiliação,
# separadas por ponto-e-vírgula (;)
# -------------------------------------------------------
COLUNA_AFILIACAO = "Affiliations"

# -------------------------------------------------------
# Função que verifica se um artigo tem afiliação do país
# Extrai o último termo de cada afiliação após a última vírgula
# -------------------------------------------------------
def contem_pais(celula, pais):
    if pd.isna(celula):
        return False
    # Divide as afiliações pelo ";"
    afiliações = str(celula).split(";")
    for afiliacao in afiliações:
        # Pega o último elemento após a última vírgula (onde fica o país)
        partes = afiliacao.strip().split(",")
        pais_encontrado = partes[-1].strip()
        if pais.lower() == pais_encontrado.lower():
            return True
    return False

# -------------------------------------------------------
# Aplica o filtro de país e intervalo de anos
# -------------------------------------------------------
df["tem_pais"] = df[COLUNA_AFILIACAO].apply(lambda x: contem_pais(x, PAIS))

df_filtrado = df[
    df["tem_pais"] &
    (df["Year"] >= ANO_INICIO) &
    (df["Year"] <= ANO_FIM)
]

total_geral  = len(df)
total_pais   = len(df_filtrado)
percentual   = (total_pais / total_geral) * 100

# -------------------------------------------------------
# Exibe o resumo geral
# -------------------------------------------------------
print(f"📊 Total de artigos no CSV          : {total_geral}")
print(f"🌍 Artigos com afiliação em {PAIS:<10}: {total_pais} ({percentual:.1f}%)")
print(f"📅 Período avaliado                 : {ANO_INICIO} — {ANO_FIM}\n")

# -------------------------------------------------------
# Exibe a contagem por ano
# -------------------------------------------------------
por_ano = df_filtrado["Year"].value_counts().sort_index()

print(f"{'Ano':<10} {'Artigos':>10}")
print("-" * 22)
for ano, qtd in por_ano.items():
    print(f"{int(ano):<10} {qtd:>10}")

print(f"\n{'TOTAL':<10} {total_pais:>10}")
```
