```python
import pandas as pd

df = pd.read_csv("meus_artigos_sla.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# Defina aqui as aplicações de interesse
# -------------------------------------------------------
aplicacoes = ["dental", "piezoelectric", "scaffold", "dielectric", "bioactive", "bioceramic", "bone"]

# -------------------------------------------------------
# Coluna onde as keywords estão armazenadas
# -------------------------------------------------------
COLUNA_KEYWORDS = "Author Keywords"

# -------------------------------------------------------
# Para cada aplicação, cria uma coluna True/False no dataframe
# indicando se aquele artigo menciona a aplicação
# -------------------------------------------------------
for aplicacao in aplicacoes:
    df[aplicacao] = df[COLUNA_KEYWORDS].str.contains(aplicacao, case=False, na=False)

# -------------------------------------------------------
# Agrupa por ano e soma as ocorrências de cada aplicação
# -------------------------------------------------------
tabela = df.groupby("Year")[aplicacoes].sum().astype(int)

# Renomeia as colunas para ficarem capitalizadas
tabela.columns = [a.capitalize() for a in aplicacoes]

# Adiciona uma coluna com o total geral por ano
tabela["Total"] = tabela.sum(axis=1)

# Adiciona uma linha com o total geral por aplicação
tabela.loc["Total"] = tabela.sum()

# -------------------------------------------------------
# Exibe a tabela no terminal
# -------------------------------------------------------
print(f"📊 Ocorrências por ano em '{COLUNA_KEYWORDS}'\n")
print(tabela.to_string())

# -------------------------------------------------------
# Salva em Excel
# -------------------------------------------------------
NOME_ARQUIVO = "aplicacoes_por_ano.xlsx"
tabela.to_excel(NOME_ARQUIVO)
print(f"\n💾 Arquivo salvo: {NOME_ARQUIVO}")
```

```python
import pandas as pd
from collections import Counter

df = pd.read_csv("meus_artigos_sla.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# Palavra-chave principal de interesse
# -------------------------------------------------------
PALAVRA_CHAVE = "scaffold"

# -------------------------------------------------------
# Frequência mínima para aparecer no resultado
# -------------------------------------------------------
FREQUENCIA_MINIMA = 3

# -------------------------------------------------------
# Coluna onde as keywords estão armazenadas
# -------------------------------------------------------
COLUNA_KEYWORDS = "Author Keywords"

# -------------------------------------------------------
# Filtra apenas os artigos que mencionam a palavra-chave
# -------------------------------------------------------
df_filtrado = df[
    df[COLUNA_KEYWORDS].str.contains(PALAVRA_CHAVE, case=False, na=False)
]

total = len(df_filtrado)
print(f"📊 Artigos com '{PALAVRA_CHAVE.capitalize()}': {total}\n")

# -------------------------------------------------------
# Extrai todas as keywords dos artigos filtrados,
# excluindo a própria palavra-chave principal
# -------------------------------------------------------
todas_keywords = []
for celula in df_filtrado[COLUNA_KEYWORDS].dropna():
    termos = [t.strip().lower() for t in str(celula).split(";") if t.strip()]
    for termo in termos:
        # Ignora a própria palavra-chave principal
        if PALAVRA_CHAVE.lower() not in termo:
            todas_keywords.append(termo)

# -------------------------------------------------------
# Conta e filtra pela frequência mínima
# -------------------------------------------------------
contagem = Counter(todas_keywords)
filtrado = [(kw, freq) for kw, freq in contagem.most_common() if freq >= FREQUENCIA_MINIMA]

# -------------------------------------------------------
# Exibe os resultados
# -------------------------------------------------------
print(f"{'Keyword':<45} {'Frequência':>10} {'% dos artigos':>15}")
print("-" * 72)
for keyword, freq in filtrado:
    percentual = (freq / total) * 100
    print(f"{keyword.capitalize():<45} {freq:>3} {percentual:>14.1f}%")

print(f"\n🔎 {len(filtrado)} keywords encontradas com frequência ≥ {FREQUENCIA_MINIMA}")
```
