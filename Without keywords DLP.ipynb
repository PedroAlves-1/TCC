# Without Keywords

```python
import pandas as pd

df = pd.read_csv("meus_artigos.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# Coluna onde as keywords estão armazenadas
# -------------------------------------------------------
COLUNA_KEYWORDS = "Author Keywords"

# -------------------------------------------------------
# Coluna onde o nome da revista está armazenado
# -------------------------------------------------------
COLUNA_REVISTA = "Source title"

# -------------------------------------------------------
# Separa artigos com e sem keywords
# -------------------------------------------------------
sem_keywords = df[
    df[COLUNA_KEYWORDS].isna() |
    df[COLUNA_KEYWORDS].str.strip().eq("")
]

com_keywords = df[
    df[COLUNA_KEYWORDS].notna() &
    ~df[COLUNA_KEYWORDS].str.strip().eq("")
]

total     = len(df)
total_sem = len(sem_keywords)

# -------------------------------------------------------
# Conta artigos por revista em cada grupo
# -------------------------------------------------------
contagem_sem = sem_keywords[COLUNA_REVISTA].value_counts().rename("Sem Keywords")
contagem_com = com_keywords[COLUNA_REVISTA].value_counts().rename("Com Keywords")
contagem_total = df[COLUNA_REVISTA].value_counts().rename("Total na Revista")

# -------------------------------------------------------
# Junta tudo em uma tabela única
# Apenas revistas que têm ao menos 1 artigo sem keyword
# -------------------------------------------------------
tabela = pd.DataFrame(contagem_sem)\
    .join(contagem_com, how="left")\
    .join(contagem_total, how="left")\
    .fillna(0)\
    .astype(int)

# Adiciona coluna de percentual de artigos sem keyword por revista
tabela["% sem KW na revista"] = (
    tabela["Sem Keywords"] / tabela["Total na Revista"] * 100
).round(1)

# Ordena pela quantidade de artigos sem keyword
tabela = tabela.sort_values("Sem Keywords", ascending=False)

# -------------------------------------------------------
# Exibe o resumo geral
# -------------------------------------------------------
print(f"📊 Total de artigos        : {total}")
print(f"❌ Sem keywords            : {total_sem} ({total_sem/total*100:.1f}%)\n")

# -------------------------------------------------------
# Exibe a tabela
# -------------------------------------------------------
print(f"{'Revista':<55} {'Sem KW':>7} {'Com KW':>7} {'Total':>7} {'% sem KW':>10}")
print("-" * 88)
for revista, linha in tabela.iterrows():
    print(
        f"{str(revista):<55} "
        f"{linha['Sem Keywords']:>7} "
        f"{linha['Com Keywords']:>7} "
        f"{linha['Total na Revista']:>7} "
        f"{linha['% sem KW na revista']:>9.1f}%"
    )

print(f"\n🔎 {len(tabela)} revistas com ao menos 1 artigo sem keyword")

# -------------------------------------------------------
# Salva em Excel
# -------------------------------------------------------
NOME_ARQUIVO = "revistas_sem_keywords.xlsx"
tabela.to_excel(NOME_ARQUIVO)
print(f"💾 Arquivo salvo: {NOME_ARQUIVO}")
```
