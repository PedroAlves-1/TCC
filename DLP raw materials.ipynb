# Materias primas

```python
import pandas as pd

df = pd.read_csv("meus_artigos.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# Defina aqui as matérias primas de interesse
# -------------------------------------------------------
materiais = ["alumina", "zirconia", "silica", "hydroxyapatite", "nitride"]

# -------------------------------------------------------
# Coluna onde as keywords estão armazenadas
# -------------------------------------------------------
COLUNA_KEYWORDS = "Author Keywords"

# -------------------------------------------------------
# Conta quantos artigos mencionam cada material
# -------------------------------------------------------
print(f"📊 Contagem de matérias primas em '{COLUNA_KEYWORDS}'")
print(f"   Total de artigos no CSV: {len(df)}\n")
print(f"{'Material':<20} {'Artigos':>10} {'% do total':>12}")
print("-" * 44)

resultados = {}
for material in materiais:
    menciona = df[COLUNA_KEYWORDS].str.contains(material, case=False, na=False)
    contagem = menciona.sum()
    percentual = (contagem / len(df)) * 100
    resultados[material] = contagem
    print(f"{material.capitalize():<20} {contagem:>10} {percentual:>11.1f}%")
```
