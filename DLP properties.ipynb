```python
import pandas as pd

df = pd.read_csv("meus_artigos.csv", encoding="utf-8-sig")

# -------------------------------------------------------
# Defina aqui as propriedades de interesse
# -------------------------------------------------------
propriedades = [
    "mechanical properties",
    "flexural",
    "compressive",
    "strength",
    "thermal conductivity",
    "porosity",
    "shrinkage",
    "microstructure",
]

# -------------------------------------------------------
# Coluna onde as keywords estão armazenadas
# -------------------------------------------------------
COLUNA_KEYWORDS = "Author Keywords"

# -------------------------------------------------------
# Para cada propriedade, cria uma coluna True/False no dataframe
# indicando se aquele artigo menciona a propriedade
# -------------------------------------------------------
for propriedade in propriedades:
    df[propriedade] = df[COLUNA_KEYWORDS].str.contains(propriedade, case=False, na=False)

# -------------------------------------------------------
# Agrupa por ano e soma as ocorrências de cada propriedade
# -------------------------------------------------------
tabela = df.groupby("Year")[propriedades].sum().astype(int)

# Renomeia as colunas para ficarem capitalizadas
tabela.columns = [p.capitalize() for p in propriedades]

# Adiciona uma coluna com o total geral por ano
tabela["Total"] = tabela.sum(axis=1)

# Adiciona uma linha com o total geral por propriedade
tabela.loc["Total"] = tabela.sum()

# -------------------------------------------------------
# Exibe a tabela no terminal
# -------------------------------------------------------
print(f"📊 Ocorrências por ano em '{COLUNA_KEYWORDS}'\n")
print(tabela.to_string())

# -------------------------------------------------------
# Salva em Excel
# -------------------------------------------------------
NOME_ARQUIVO = "propriedades_por_ano.xlsx"
tabela.to_excel(NOME_ARQUIVO)
print(f"\n💾 Arquivo salvo: {NOME_ARQUIVO}")
```
